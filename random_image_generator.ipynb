```python
'''Generate random gray level image'''
import numpy as np
from random import *
import matplotlib.pyplot as plt


size_of_image=[200, 300]
#generate a random gray level image
matrix = np.zeros((size_of_image[0],  size_of_image[1]))


for i in range(1,len(matrix)):
    for j in range(1, len(matrix[0])):
        matrix[i][j]=randint(0, 255)
      
plt.imshow(matrix, cmap='gray')
plt.show()
        
        
matrix_new = np.zeros((size_of_image[0],  size_of_image[1]))
#generate a random gray level image where the values of some pixels have some dependance to their neighbors.
for i in range(1,len(matrix_new)):
    for j in range(1, len(matrix_new[0])):
        if random() <0.9:
            matrix_new[i][j]=0.9*matrix_new[i-1][j]+0.1*matrix_new[i][j-1]
        else:
            matrix_new[i][j]=randint(0, 255)
#for i in range(100):
    #print(randint(0, 255))

plt.imshow(matrix_new, cmap='gray')
plt.show()
```
